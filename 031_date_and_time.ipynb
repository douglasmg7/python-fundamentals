## Date and Time

### The time module

```python
# https://docs.python.org/3/library/timeit.html
```

```python
from time import perf_counter, sleep
```

```python
# perf_counter, count seconds from when the program start.
start = perf_counter()
start
```

```python
end = perf_counter()
```

```python
# Give the elapsed time in seconds.
end - start
```

```python
# Cheking sleep time.
start = perf_counter()
sleep(30)
end = perf_counter()
elapsed_time = end - start
print(elapsed_time)
```

```python
# Using strftime
from time import strftime
```

```python
now = gmtime(time())
now
```

```python
strftime('%Y/%m/%d', now)
```

```python
# Parsing string to struct time (string parse time).
from time import strptime
```

```python
strptime('2026/05/02', '%Y/%m/%d')
```

```python
s = 'Monday, April 18, in the year 2020 CE'
```

```python
fmt = '%A, %B %d, in the year %Y CE'
```

```python
strptime(s, fmt)
```

```python
# %Y - Four digits year
# %y - Two digits year

# %m - Month number
# %B - Month full name
# %b - Month abbreviated name

# %d - Day of month number

# %H - Hour in 24-hour clock
# %I - Hour in 12-hour clock
# %p - AM or PM

# %M - Minute number
# %S - Second number

# %z - Time zone offset +-HHMM
# %Z - Time zone name

# %w - Weekday number (sunday=0)
# %A - Weekday full name
# %a - Weekday abbreviated name
```

### The datetime module

```python
import datetime
```

```python
dt = datetime.date(2026, 1, 3)
dt
```

```python
# Local time, not utc.
dt = datetime.date.today()
dt
```

```python
import time
```

```python
ep = time.time()
ep
```

```python
dt = datetime.date.fromtimestamp(ep)
dt
```

```python
dt = datetime.date.fromisoformat('2020-02-01')
dt
```

```python
dt = datetime.date(2026, 3, 1)
dt
```

```python
dt.isoformat()
```

```python
dt.year, dt.month, dt.day
```

```python
t = datetime.time(15, 30, 12)
t
```

```python
t = datetime.time(15, 30, 12, 1023)
t
```

```python
t.isoformat()
```

```python
t.hour, t.minute, t.second, t.microsecond
```

```python
t = datetime.time.fromisoformat('23:15:03.234')
t
```

```python
dt = datetime.datetime(2026, 3, 1, 13, 12, 15, 123456)
dt
```

```python
dt.isoformat()
```

```python
dt = datetime.datetime.fromisoformat('2026-03-01T13:12:15.123456')
dt
```
