## Iterables and iterators
A iterable can be used with a iter() to get a iterator that can be used if next()

```python
l = [1, 2, 3, 4, 5]
for el in l:
    print(el)
print('Finish')
```

```python
l = [1, 2, 3, 4, 5]

iterator = iter(l)
print(type(iterator))
try:
    while True:
        print(next(iterator))
except StopIteration:
    pass
print('Finish')
```

```python
l = [1, 2]

iterator = iter(l)
print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))
```

```python
# Range is lazy iterable.
r = range(1,4)
iterate = iter(r)
try:
    while True:
        print(next(iterate))
except StopIteration:
    pass
    
```

```python
# Range is lazy iterable.

# Create a iterable.
r = range(1,10)

# Create a iterator.
r_iterate = iter(r)

print(next(r_iterate))
print(next(r_iterate))

l = list(r_iterate)
print(l)
```

```python
from time import perf_counter

start = perf_counter()
# Not create the whole list.
l = range(1, 100_000_000)
print(f'Elapsed time: {perf_counter() - start}')
```

```python
from time import perf_counter

start = perf_counter()
# Create the whole list.
l = list(range(1, 100_000_000))
print(f'Elapsed time: {perf_counter() - start}')
del l
```
