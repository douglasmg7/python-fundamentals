### if else

```python
a = 1
b = 2
if a > b:
    print('a is bigger')
else:
    print('b is bigger')
```

### if elif else

```python
a = 5000
b = 5000
if a > b:
    print('a is bigger')
elif a == b:
    print('a equal b')
else:
    print('b is bigger')
```

### Ternary operator

```python
x=1
print('positive') if x>0 else print('negative')
```
