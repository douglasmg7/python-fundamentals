### Map

```python
data = ['a', 'ab', 'abc', 'abcd']
```

```python
lengths = [len(el) for el in data]
lengths
```

```python
lengths = (len(el) for el in data)
```

```python
lengths
```

```python
3 in lengths
```

```python
list(lengths)
```

```python
# map is a cleaner syntax.
lengths = map(len, data)
```

```python
lengths
```

```python
list(lengths)
```
