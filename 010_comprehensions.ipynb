### Comprehensions
Comprehension are an easy way to create new iterables from other iterables.  
Comprehension create all the items.  
Generators create items on demand.

```python
numbers = [1, 2, 3, 4, 5, 6]
numbers
```

```python
square = list()
for n in numbers:
    square.append(n**2)
square
```

```python
square = [number**2 for number in numbers]
square
```

```python
evens = [number**2 for number in numbers if number % 2 == 0]
evens
```
