### Sorting

```python
data = [-10, -6, 0, 3, 6]
sorted(data, key=abs)
```

```python
letters = ['a', 'z', 'A', 'X', 'Z', 'x']
sorted(letters)
```

```python
sorted(letters, key=lambda k: k.casefold())
```

```python
text = 'era uma vez um grande terremoto em algum lugar'.split()
```

```python
sorted(text, key=lambda el: len(el))
```
