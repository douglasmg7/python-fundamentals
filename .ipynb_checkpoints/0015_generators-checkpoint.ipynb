# Generators

```python
# It is not a tuple comprehension but a Generators.
# Generators are iterators.
# Generators calculate and hand out elements one at a time as request.
ge = (el ** 2 for el in range(1, 8))
print(type(ge))
print(ge)
```

```python
print(next(ge))
print(next(ge))
print(next(ge))
```

```python
for i in ge:
    print(i)
```

```python
ge = (el ** 2 for el in range(1, 8))
# ge is a generator alredy. Calling iter(), return her self.
iter(ge) == ge
```

```python
# Be aware some times generator is consumed.
ge = (el ** 2 for el in range(1, 8))
print(25 in ge)
list(ge)
```

```python
# Comparing time of creation of comprehension and generators.
from time import perf_counter
# Generator.
start = perf_counter()
ge = (el ** 2 for el in range(1, 10_000_000))
print(f'Generator: {perf_counter() - start}')
# Comprehension.
start = perf_counter()
ge = [el ** 2 for el in range(1, 10_000_000)]
print(f'Comprehension: {perf_counter() - start}')
```
