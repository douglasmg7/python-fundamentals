### Comprehensions
Comprehension are an easy way to create new iterables from other iterables.  
Comprehension create all the items.  
Generators create items on demand.

```python
numbers = [1, 2, 3, 4, 5, 6]
numbers
```

```python
square = list()
for n in numbers:
    square.append(n**2)
square
```

```python
square = [number**2 for number in numbers]
square
```

```python
evens = [number**2 for number in numbers if number % 2 == 0]
evens
```

```python
sales = {"s1": 4, "s2": 7, "s3": 1, "s4": 5, "s5": 9}
```

```python
high_sales = list()
for key, value in sales.items():
    if value > 4:
        high_sales.append(key)
high_sales
```

```python
low_sales = [key for key, value in sales.items() if value < 7]
low_sales
```

### Dictionary comprehesion

```python
widgets = ["widget 1", "widget 2", "widget 3", "widget 4"]
sales = [10, 5, 15, 0]
d = {}
for i in range(len(widgets)):
    if sales[i] > 0:
        d[widgets[i]] = sales[i]
d
```

```python
widgets = ["widget 1", "widget 2", "widget 3", "widget 4"]
sales = [10, 5, 15, 0]
d = { widgets[i]: sales[i] for i in range(len(widgets)) if sales[i] > 0 }
d
```

```python
### Count words repetitions.
```

```python
words  = text.split(" ")
```

```python
text_no_spaces = "".join(words).lower()
```

```python
freq = {}
for char in set(text_no_spaces):
    freq[char] = len([ item for item in text_no_spaces if item == char])
dict(sorted(freq.items()))
```

```python
dict(sorted(freq.items(), key=lambda item: item[1]))
```

### Using only comprehend.

```python
freq = {
    element: len([char for char in text_no_spaces if char == element])
    for element in set(text_no_spaces) 
}
dict(sorted(freq.items()))
```
