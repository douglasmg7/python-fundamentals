### Passing and returnning functions

```python
def add(a, b):
    return a + b

def greet(name):
    return f'Hello {name}'
```

```python
add(2,3), greet('Samanta')
```

```python
def apply(func, *args):
    return func(*args)
```

```python
apply(add, 1, 2)
```

```python
apply(greet, 'Juliana')
```

```python
apply(greet, 'Juliana', 'Marcelo')
```

```python
apply(lambda a, b, c: a + b + c, 10 ,20 ,30)
```

```python
def mult(a, b):
    return a * b

def power(a, n):
    return a ** n
```

```python
def choose_operator(name):
    if name == 'add':
        return add
    elif name == 'mult':
        return mult
    else:
        return power
```

```python
choose_operator('add')(4, 5)
```

```python
choose_operator('add')
```

```python
def choose_operator(name):

    def add(a, b):
        return a + b

    def power(a, n):
        return a ** n

    def mult(a, b):
        return a * b
    
    if name == 'add':
        return add
    elif name == 'mult':
        return mult
    else:
        return power
```

```python
choose_operator('power')(2, 8)
```

```python
choose_operator('add')
```

```python
def choose_operator(name):  
    if name == 'add':
        return lambda a, b: a + b
    elif name == 'mult':
        return lambda a, b: a * b
    else:
        return lambda a, b: a ** b
```

```python
choose_operator('power')(2, 10)
```

```python
choose_operator('power')
```
