## Date and Time

### Time module

```python
# https://docs.python.org/3/library/timeit.html
```

```python
from time import perf_counter, sleep
```

```python
# perf_counter, count seconds from when the program start.
start = perf_counter()
start
```

```python
end = perf_counter()
```

```python
# Give the elapsed time in seconds.
end - start
```

```python
# Cheking sleep time.
start = perf_counter()
sleep(30)
end = perf_counter()
elapsed_time = end - start
print(elapsed_time)
```

```python
from time import gmtime
```

```python
# Epoch start time.
gmtime(0)
```

```python
# Time from seconds from epoc time.
gmtime(1_000_000_000)
```

```python
gmtime(-1_000_000_000)
```

```python
# Tomorrow.
gmtime(time() + (24 * 60 * 60))
```

```python
# Convert a time struct to a epoch time.
from calendar import timegm
```

```python
current = time()
current
```

```python
current = gmtime(current)
current
```

```python
# Will lost decimal seconds, because time sctruct no use it.
timegm(current)
```

```python
# Using strftime
from time import strftime
```

```python
now = gmtime(time())
now
```

```python
strftime('%Y/%m/%d', now)
```

```python
# Parsing string to struct time (string parse time).
from time import strptime
```

```python
strptime('2026/05/02', '%Y/%m/%d')
```

```python
s = 'Monday, April 18, in the year 2020 CE'
```

```python
fmt = '%A, %B %d, in the year %Y CE'
```

```python
strptime(s, fmt)
```
