### Closures

```python
def outer(a, b):
    sum_ = a + b

    def inner():
        prod = a * b
        print(a, b, sum_, prod)
        print('You just called a closure!')

    return inner
```

```python
f = outer(40, 50)
```

```python
f
```

```python
f()
```

```python
f.__closure__
```

```python
def power(n):
    def inner(x):
        return x ** n

    return inner
```

```python
square = power(2)
cube = power(3)
```

```python
square(5)
```

```python
cube(5)
```

```python
square.__closure__
```

```python
cube.__closure__
```

```python
def execute(func):
    def inner(*args, **kwargs):
        return func(*args, **kwargs)

    return inner
```

```python
def add(a, b, c):
    print('add...')
    return a + b + c
```

```python
def say_hello(name, *, formal=True):
    print('say_hello...')
    if formal:
        print(f'Pleased to meet you, {name}')
    else:
        print(f'Hi, {name}')
```

```python
exec_add = execute(add)
exec_say_hello = execute(say_hello)
```

```python
exec_add(10, 20, 30)
```

```python
exec_say_hello('Fred', formal=False)
```
