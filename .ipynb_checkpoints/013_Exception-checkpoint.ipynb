```python
ex = KeyError('Some message')
```

```python
isinstance(ex, LookupError), isinstance(ex, Exception), isinstance(ex, IndexError)
```

```python
# Log a error message an raise the same exception again.
try:
    r = 1/0
except Exception as ex:
    print(f'Loggin the exception: {ex}')
    raise
# Always run.
finally:
    print('Bye bye!')
```

```python
try:
    r = 1/0
except ZeroDivisionError as ex:
    print(f'Exception ocurred: {type(ex)}, {ex}')
finally:
    print('Finally runs')
print('The code continued')
```

```python
# Using exception to loop.
l = [1, 2, 3]

while True:
    try:
        print(l.pop())
    except IndexError:
    # IndexError means empty list.
        print('All done.')
        break
print('Code resumes here.')
```

```python
# Can use try without except.
try:
    print('Nothing here')
    1/0
finally:
    print('Alaways run')
```
