### Reshaping Arrays

```python
m = np.arange(12)
m, m.shape
```

```python
m1 = m.reshape(3, 4)
m1, m1.shape
```

```python
m2 = m1.reshape(2, 6)
m2, m2.shape
```

```python
m3 = m2.reshape(12,)
m3
```

```python
m[0] = 100
```

```python
m, m1, m2
```

```python
m1[1][2] = 60
```

```python
m
```

```python
m4 = m.reshape(4, 3).copy()
```

```python
m4[0][0] = -1
```

```python
m4, m
```

### Stacking Arrays

```python
a1 = np.arange(6,11)
a1
```

```python
a2 = np.arange(1, 11).reshape(2, 5)
a2
```

```python
v1 = np.vstack((a1, a2))
v1
```

```python
a1 = np.linspace(1, 5, 6).reshape(3, 2)
a1
```

```python
a2 = np.random.randint(1, 10 ,6).reshape(3, 2)
a2
```

```python
h1 = np.hstack((a1, a2))
h1
```

### Slicing

```python
ar = np.arange(1,7)
ar
```

```python
ar2 = ar[:3]
ar2
```

```python
# The arrarys are linked, differente from list.
ar2[0] = 100
ar2 ,ar
```

```python
ar = np.arange(1,26).reshape(5, 5)
ar
```

```python
# The first parameter is index 0 and the second is index 1 (column).
ar[:2, :2]
```

```python
ar[::2, ::2]
```

```python
ar[::2, ::2] = 100
ar
```

```python
ar = np.arange(1,26).reshape(5, 5)
ar
```

```python
ar[2, 1::2]
```

```python
ar = np.arange(1,7)
ar
```

```python
ar[:3] = np.array([100, 200, 300])
ar
```

```python
# Can use list too.
ar[:3] = [400, 500, 600]
ar
```

```python
# Not work, must have the same size.
ar[:3] = np.array([100, 200, 300, 400])
```

```python
ar = np.arange(9).reshape(3,3)
ar
```

```python
ar[::2, ::2]
```

```python
ar[::2, ::2] = [[100, 200], [600, 800]]
ar
```

```python
# Using broadcasting with scalar value.
ar[::2, ::2] = 700
ar
```
