# Functions

```python
# Parameter is when we define the function
# Argument is when we call de function
```

```python
# Function are Python objects
# They have state
#    name
#    code
#    parameter
# They are callable, and always return something when called
# They can be assigned to a symbol
# Can be passed as a parameter to another function
```

```python
# If a function do not have a return value, Python will make it return None object
def my_func():
    print('Hello!')
print(my_func())
if my_func() == None:
    print('Nothing to do')
```

```python
# Every time a function is called it create a dict (local) into her namespace to
# keep all arguments. key is the parameter name and value is the argument when called.
```

```python
def say_hello():
    return 'Hello!'
```

```python
globals()
```

```python
f = say_hello
```

```python
id(f), id(say_hello), f is say_hello
```

```python
f()
```

```python
del say_hello
```

```python
say_hello()
```

```python
say_hello.__name__, f.__name__
```

```python
def sum_numbers(a, b):
    sum = a + b
    print(locals())
    return sum
```

```python
sum_numbers(4, 5)
```

```python
def matrix(rows, cols, default):
    return [[default for i in range(cols)] for j in range(rows)]
```

```python
matrix(4, 8, 1)
```

```python
matrix(default=3, cols=8, rows=4)
```

```python
matrix
```

### Start argument

```python
def mfunc(*args):
    print(type(args))
    print(args)
    for el in args:
        print(el)
```

```python
mfunc(1, 2, 3)
```

```python
l = [1, 2, 3, 4]
a, b, c, d = l
a, b, c, d
```

```python
# Unpack
mfunc(*l)
```

```python
# Default value
def my_func(a=1, b=2, c=3):
    print(locals())
my_func(5)
```

```python
# Once we have default arguments we need specify default for all parameters after it
def my_func(a, b, c=3, d=4, *args):
    print(locals())
my_func(1, 2, 5, 6, 7, 8, 9)
```
