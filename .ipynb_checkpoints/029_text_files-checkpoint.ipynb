```python
try:
    file = open('simple_sample.csv')
    print(next(file))
    print(next(file))
    print(next(file))
    print(next(file))
    print(next(file))
except Exception as ex:
    print(f'Loggin the exception: {ex}')
    raise
finally:
    print('Closing the file...')
    file.close()
```

```python
# Using context manage:
```

```python
with open('simple_sample.csv') as file:
    data = file.readlines()
    print(file.closed)
print(file.closed)
data    
```

```python
with open('simple_sample.csv') as file:
    header = file.readline()
    for line in file:
        line = line.strip()
        data, v1, v2, v3 = line.split(',')
        print(data, v1, v2, v3)
```

```python
data = ['line 1\n', 'line 2\n', 'line 3\n']
with open('test.csv', 'w') as file:
    file.writelines(data)
```

```python
with open('test.csv', 'r') as file:
    lines = file.readlines()
lines
```

```python
with open('test.csv', 'r') as file:
    for line in file:
        print(line)
```

```python
with open('test.csv', 'r') as file:
    for line in file:
        print(line, end='')
```

```python
with open('test.csv', 'a') as file:
    file.write('line 4\n')
```
