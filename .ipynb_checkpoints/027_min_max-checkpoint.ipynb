### min and max

```python
data = [-1, -4, 3, 8, -9]
```

```python
min(data)
```

```python
max(data)
```

```python
min(data, key=abs)
```

```python
max(data, key=abs)
```

```python
data = {
    'S1': (100, 200, 80, 180),
    'S2': (10, 20, 8, 18),
    'S3': (50, 150, 50, 150)
}
```

```python
min(data, key=lambda key: data[key][1] - data[key][2])
```
