```python
data = [
    ('item_1', 10, 100.0),
    ('item_2', 5, 25.0),
    ('item_3', 100, 0.25)
]
```

```python
schema = ('widget', 'num_sold', 'unit_price')
```

```python
d = {}
for item in data:
    d[item[0]] = {'num_sold': item[1], 'unit_price': item[2]}
```

```python
d
```

```python
schema = ('widget', 'num_sold', 'unit_price')
```

```python
data
```

```python
for row in data:
    print(list(zip(schema, row)))
```

```python
result = dict()
for row in data:
    result[row[0]] = dict(zip(schema[1:], row[1:]))
result
```

```python
# Using conprehension.
{ row[0]: dict(zip(schema[1:], row[1:])) for row in data }
```

```python
data = [
    ('item_1', 'Ford', 10, 100.0, .1),
    ('item_2', 'GM', 5, 25.0, .2),
    ('item_3', 'Fiat', 100, 0.25, .3)
]
```

```python
schema = ('widget', 'Manufacturer', 'num_sold', 'unit_price', 'discount')
```

```python
r = { row[0]: dict(zip(schema[1:], row[1:])) for row in data }
r
```
