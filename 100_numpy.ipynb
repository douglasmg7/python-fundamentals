### Reshaping Arrays

```python
m = np.arange(12)
m, m.shape
```

```python
m1 = m.reshape(3, 4)
m1, m1.shape
```

```python
m2 = m1.reshape(2, 6)
m2, m2.shape
```

```python
m3 = m2.reshape(12,)
m3
```

```python
m[0] = 100
```

```python
m, m1, m2
```

```python
m1[1][2] = 60
```

```python
m
```

```python
m4 = m.reshape(4, 3).copy()
```

```python
m4[0][0] = -1
```

```python
m4, m
```
