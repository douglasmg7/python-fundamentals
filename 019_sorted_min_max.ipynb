Sorted, min, max

```python
l = [1, 10, 2, 9, 3, 8]
t = (1, 10, 2, 9, 3, 8)
s = {1, 10, 2, 9, 3, 8}
```

```python
sorted(l), sorted(t), sorted(s)
```

```python
min(l), min(t), min(s)
```

```python
max(l), max(t), max(s)
```

```python
sorted(l, reverse=True), sorted(t, reverse=True), sorted(s, reverse=True)
```
