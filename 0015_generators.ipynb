# Generators

```python
# It is not a tuple comprehension but a Generators.
# Generators are iterators.
# Generators calculate and hand out elements one at a time as request.
ge = (el ** 2 for el in range(1, 5))
print(type(ge))
print(ge)
```

```python
print(next(ge))
print(next(ge))
print(next(ge))
```
