## Lambda functions

```python
# Lambda function is a expression that returns a function object.
# It not assigned to an symbol(name).
# Also called anonymous function.
f = lambda a, b: a + b
f(1, 2)
```

```python
# Identity matrix
f = lambda rows, cols: [
    [1 if row == col else 0 for col in range(cols)]
    for row in range(rows)
]
```

```python
f(5,5)
```
