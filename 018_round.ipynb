### Round() built-in

```python
round(.325)
```

```python
round(.875)
```

```python
# Round to closest even number.
round(13.5), round(12.5)
```

```python
round(.125), round(.125, 0), round(.125, 1), round(.125, 2), round(.125, 3)
```

```python
round(123456, -1), round(123456, -2), round(123456, -3), round(123456, -4)
```

```python
format(.325, '.20f')
```

```python
print(f'{.325:.20f}')
```
